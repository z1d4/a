# 8-b Generate similar sentences from a given Hindi text input 

```python
pip install torch==1.3.1
```

```python
from inltk.inltk import setup
setup('hi')
from inltk.inltk import get_similar_sentences
output= get_similar_sentences("मैं बहुत खुश हूं।",5,'hi')
print(output)
```

```python
from inltk.inltk import get_similar_sentences

text = 'गीक्स फॉर गीक्स एक बेहतरीन टेक्नोलॉजी लर्निंग प्लेटफॉर्म है।'
n=5
get_similar_sentences(text, n, 'hi')

```
