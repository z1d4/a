```python
from nltk.corpus import brown
import nltk
nltk.download('brown')
brown.words()
```

```python
from nltk.corpus import brown
#nltk.download('brown')
brown.categories()
```

```python
news_text=brown.words(categories='news')
fdist=nltk.FreqDist([w.lower() for w in news_text])
modals=['can','could','might','must','will']
for m in modals:
  print(m + ':', fdist[m])
```

```python
from nltk.inference.tableau import Categories
cfd=nltk.ConditionalFreqDist((genre,word)
for genre in brown.categories()
for word in brown.words(categories=genre))
genres=['news','religion','hobbies','science_fiction','romance','humor']
cfd.tabulate(conditions=genres,samples=modals)
```

```python
coconut_text=reuters.words(categories='coconut')
fdist=nltk.FreqDist([w.lower() for w in coconut_text])
modals=['can','could','might','must','will']
for m in modals:
  print(m + ':', fdist[m])
```

```python
from nltk.inference.tableau import Categories
cfd=nltk.ConditionalFreqDist((genre,word)
for genre in reuters.categories()
for word in reuters.words(categories=genre))
genres=['coconut','job','zinc','barley','coca','corn']
cfd.tabulate(conditions=genres,samples=modals)
```

```python
from nltk.corpus import reuters
cfd=nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in reuters.fileids()
    for w in reuters.words(fileid)
    for target in ['barley','corn']
    if w.lower().startswith(target))
cfd.plot()
```

udhr conditional probability distribution

```python
import nltk
nltk.download('udhr')
```

```python
import nltk
from nltk.corpus import udhr
languages=['Chickasaw','English','German_Deutsch']
cfd=nltk.ConditionalFreqDist(
    (lang, len(word))
    for lang in languages
    for word in udhr.words(lang + '-Latin1'))
cfd.plot(cumulative=True)
```
