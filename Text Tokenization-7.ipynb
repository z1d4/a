```python
#prac7-a Tokenization using python split function
text="this tool is in the beta stage.Alex developers can use metrics"
data=text.split('.')
for i in data:
  print(i)
```

```python
#7-b Tokenization using regular expression
import nltk
from nltk.tokenize import RegexpTokenizer
tk=RegexpTokenizer('s\+',gaps=True)
str="I love to study Natural Language Pocessing in Python"
tokens=tk.tokenize(str)
print (tokens)
```

```python

import nltk
from nltk.tokenize import word_tokenize
str="I love to study Natural Language Pocessing in Python"
print(word_tokenize(str))
```

```python
#7-d Tokenization using spacing library
import spacy
nlp= spacy.blank("en")
str="I love to study Natural Language Processing in Python"
doc= nlp(str)
words=[word.text for word in doc]
print(words)
```
